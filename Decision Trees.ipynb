# Decision Trees

A simple summary and exmaple of how decision trees work!


A decision tree basicaly has nodes where the model decides to split outcomes on a feature at a certain threshold. In the above example, at this particular node, if Age is < 56, it splits one way, and > 56 it splits another way.

When they are being built decision trees are constructed by recursively evaluating different features and using at each node the feature that best splits the data. This is done by evaluating certain metrics, like the Gini index or the Entropy for categorical decision trees, or the Residual or Mean Squared Error for regression trees.

```python
import pandas as pd
data = pd.read_csv('credit_card.csv')
data
```

```python
X = data.loc[:,["Purchases","SuppCard"]]
y = data.loc[:,["Upgraded"]]
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gbc = GradientBoostingClassifier()
```

```python
gbc.fit(X,y)
```

```python
pred = gbc.predict(X)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y,pred))
```

```python
import joblib
joblib.dump(gbc,'credit_card_model')
```
